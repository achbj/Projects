```python
import pandas as pd
from pycaret.regression import *
```

```python
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')
```

```python
rgn = setup(train, target="SalePrice", ignore_features=["ID"])
best_model = compare_models(sort='RMSE')
```

```python
# save_model(best_model, model_name='rgn_model_ab90')
```

```python
final_model = finalize_model(best_model)
```

```python
test_predictions = predict_model(final_model, data=test)
submission = test_predictions[['Id', 'prediction_label']]
submission.columns = ['Id', 'SalePrice']
submission.to_csv('submission_4.csv', index=False)
```

```python
tuned_model = tune_model(best_model,search_library='optuna' ,optimize='RMSE', choose_better=True)
```
