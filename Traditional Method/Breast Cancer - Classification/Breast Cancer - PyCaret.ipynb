```python
import pandas as pd
from pycaret.classification import *
```

## Data source:
kaggle [link](https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data)

```python
df = pd.read_csv('./breast_cancer_prediction.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

## PyCaret Setup

```python
clf = setup(df, target='diagnosis', ignore_features=['id','Unnamed: 32'])
best_model = compare_models()
```

```python
lr = create_model('lr')
```

## Tune Model

```python
tuned_lr = tune_model(lr)
```

## Finalize Model

```python
final_model = finalize_model(tuned_lr)
```

## Make Prediction

```python
# prediction on Validation Set seperated by pycaret
predict_model(final_model)
```

```python
evaluate_model(final_model)
```
