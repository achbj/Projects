# Heart Disease Detection - `PyCaret` and `mlflow`

## Import Libraries

```python
import pandas as pd
from pycaret.classification import *
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./heart.csv')
df.head()
```

```python
df.info()
```

## Training and Test Set Split

```python
# Separate features and target
X = df.drop(["target"], axis=1)
y = df['target']
```

```python
# train and test set split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

## Pycaret Setup

```python
clf = setup(data=X_train, target=y_train, log_experiment=True, experiment_name='HeaertDiseaseEXP_1')
best_model = compare_models()
```

```python
lr_model = create_model('lr', experiment_custom_tags='exp_2')
```

```python
tuned_model = tune_model(lr_model)
```

```python
plot_model(lr_model, plot='confusion_matrix', plot_kwargs={'percent': True})
```

```python
tuned_model_1 = tune_model(lr_model, optimize='Accuracy')
```

```python
plot_model(tuned_model_1, plot='confusion_matrix', plot_kwargs={'percent': True})
```

```python
tuned_model_2 = tune_model(lr_model, n_iter=100, optimize="Accuracy", choose_better=True)
```

```python
tuned_model_2 = tune_model(lr_model, n_iter=1000, optimize="Accuracy", choose_better=True)
```

```python
plot_model(tuned_model_2, plot='vc')
```

```python
save_model(tuned_model_2, model_name='HDD-86')
```

```python
final_model = load_model('HDD-86')
```

```python
y_pred = final_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

## Conclusion  
From Using the pycaret we have found that mostly `Extreme Gradient Boosting`, `Extra Trees Classifier` provides with result in overal score. Using logistic regression highest we have reached is `86.03%` on training set while `78.5%` in test set.

If we want to use logistic regression in for this project, we will get better result with sklearn library as it gives `85.85%` on training set and `81.46%` on test set. Both pycaret and sklearn models are not over fitting as our evaluation of train test curve. 

With multiple experiment we can see that `xgboost` is always above 95% of accuracy while other keep moving up and down. If we want to get better result with `logistic regression` than we can play around with features and other hyperparameter but then there is more likely to be over fitting or other problems to face.
