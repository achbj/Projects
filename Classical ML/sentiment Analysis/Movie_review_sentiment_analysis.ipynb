```python
import nltk
import random
from nltk.corpus import movie_reviews
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
nltk.download('movie_reviews')
```

```python
movie_reviews.fileids()
```

```python
documents = [(list(movie_reviews.words(fileid)), category)
 for category in movie_reviews.categories()
 for fileid in movie_reviews.fileids(category)]

# shuffle the data to ensure random distribution of positive and negative review
random.shuffle(documents)

# Separate the documents and labels
documents = [(" ".join(words), label) for words, label in documents]
documents, labels = zip(*documents)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(documents, labels, test_size=0.2, random_state=42)
```

```python
# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the training data
X_train_vectors = vectorizer.fit_transform(X_train)

# Transform the testing data using the same vectorizer
X_test_vectors = vectorizer.transform(X_test)
```

```python
# Initialize the SVM classifier
classifier = SVC(kernel='linear', C=1.0)

# Train the classifier on the training data
classifier.fit(X_train_vectors, y_train)
```

```python
# Make predictions on the testing data
y_pred = classifier.predict(X_test_vectors)

# Evaluate the model's accuracy and other metrics
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:")
print(report)
```
