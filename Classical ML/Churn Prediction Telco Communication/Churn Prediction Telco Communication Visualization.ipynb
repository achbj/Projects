# Data Visualization

Let's visualize the ratio of the churned and not churned users ratio first and their ratio to other features

```python
# Count the number of churned and non-churned customers
churn_counts = df_en[['Churn_Yes', 'Churn_No']].sum()

# Labels for the pie chart
labels = ['Churned', 'Not Churned']

# Colors for the piechart
colors = ['red', 'blue']

# Explode a slice
explode = (0.1, 0)

# Create a piechart
plt.figure(figsize=(6,6))
plt.pie(churn_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)

# Add a legend in the right top corner
plt.legend(labels, loc="upper right")

plt.title('Churn Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the total number of customers
total_customers = len(df_en)

# Calculate the number of churned and not churned customers by gender
churn_counts = df_en.groupby('gender_Male')[['Churn_Yes', 'Churn_No']].sum()

# Calculate the gender ratio (percentage)
gender_ratio = df_en['gender_Male'].value_counts(normalize=True) * 100

# Labels for the pie chart
labels = ['Churned (Female)', 'Not Churned (Female)', 'Churned (Male)', 'Not Churned (Male)']

# Colors for the pie chart based on gender ratio
colors = ['red', 'blue', 'red', 'blue']

# Explode slices if needed (e.g., to emphasize 'Churned')
explode = (0.1, 0, 0.1, 0)

# Create a pie chart
plt.figure(figsize=(6,6))
plt.pie(churn_counts.values.flatten(), labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)


# Title
plt.title('Churn Distribution by Gender')
plt.show()
```

From the above piechart we can see that churned and Not churned have similar ratio in relation to gender

```python
# Create correlation matrix
corr_matrix = df_en.corr()

plt.figure(figsize=(15, 15))
sns.clustermap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

There are too many columns and it is very difficult to visualize here so let's create a threshold of 10% so only the column with above 10% of the correlation or below 10% of correlation can be selected

```python
# Set the correlation threshold
threshold = 0.10

# Filter the correlation matrix to include only corr above threshold
high_corr_mat = corr_matrix[(corr_matrix > threshold) | (corr_matrix < -threshold)]
plt.figure(figsize=(15, 10))
sns.heatmap(high_corr_mat, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title(f'Correlation Heatmap (Threshold > {threshold})')
plt.show()
```

```python
# Define the stages and the number of customers at each stage
stages = ["Total Customers", "Customers with Contracts", "Customers with Internet Service", "Churned Customers"]
customer_counts = [
    len(df_en),  # Total Customers
    len(df_en[df_en['Contract_Month-to-month'] == 1]),  # Customers with Contracts
    len(df_en[df_en['InternetService_No'] == 0]),  # Customers with Internet Service
    len(df_en[df_en['Churn_Yes'] == 1]),  # Churned Customers
]

# Create a funnel chart
plt.figure(figsize=(8, 6))
plt.barh(stages, customer_counts, color='skyblue')
plt.xlabel('Number of Customers')
plt.title('Churn Funnel Chart')
plt.gca().invert_yaxis()  # Invert the y-axis to display the funnel from top to bottom
plt.show()
```
